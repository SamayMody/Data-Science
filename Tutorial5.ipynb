```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# This dataset is about KALYAN JWELLERS script listed in the NSE and BSE
# It includes the Daily price details of this script from 5/9/24 to 4/10/24
df = pd.read_csv('KALN Historical Data.csv')
df
```

```python
# Line plot
plt.plot(df['Date'], df['Price'])
plt.title('Close Price')
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.show()
```

```python
plt.plot(df['Date'], df['High'], label = 'High')
plt.plot(df['Date'], df['Low'], label = 'Low')
plt.plot(df['Date'], df['Open'], label = 'Open')
plt.plot(df['Date'], df['Price'], label = 'Close')
plt.title('Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('KALYAN JWELLERS')
plt.legend()
plt.show()
```

```python
# vertical bar
plt.bar(df['Date'], df['Vol.'])
plt.xlabel('Date')
plt.ylabel('Vol')
plt.title('Volume')
plt.show()
```

```python
# horizontal bar
plt.barh(df['Date'], df['Vol.'])
plt.xlabel('Volume')
plt.ylabel('Date')
plt.title('Open Price')
plt.show()
```

```python
plt.scatter(df['Open'], df['Price'])
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Open vs Close')
plt.show()
```

```python
# box plot
plt.boxplot(df['Price'])
plt.xlabel('KALYANKJIL')
plt.ylabel('Close')
plt.title('Open Price')
plt.show()
```

```python
sns.histplot(df['Price'])
plt.xlabel('Price')
plt.ylabel('Num')
plt.title('Closing Price')
plt.show()
```
